```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Exploratory Data Analysis (EDA)

EDA is a critical step in understanding the structure and behaviour of the dataset before building insights or models. It helps to summarize the main characteristics of the data, detect patterns, identify anomalies, and understand relationships between variables.

In this project, the social media marketing dataset was explored to understand user behaviour, platform performance, campaign effectiveness, and engagement trends using the following steps:

Generated summary statistics using describe() to understand the overall distribution of numerical variables such as engagement and followers.

Performed univariate analysis to examine individual features including platform distribution, campaign type, industry, age group, and engagement levels.

Conducted bivariate analysis to study relationships between variables such as platform vs engagement, followers vs engagement, and campaign type vs engagement.

Applied multivariate analysis to analyze the combined effect of multiple variables such as platform, industry, campaign type, followers, and engagement.

Used group-by operations to calculate average engagement and followers across different platforms and campaign types.

Created aggregated comparisons to analyze performance trends across platforms and industries.

Performed correlation analysis to identify relationships between numerical variables in the dataset.

Used a correlation heatmap to visually represent the strength and direction of relationships between numerical features.

This EDA process helped uncover meaningful insights into social media marketing performance and provided a strong foundation for data-driven decision-making.

```python
df.describe()
```

```python
plt.figure()
sns.countplot(x='platform', data=df, hue='platform', legend=False)
plt.title("Platform Distribution")
plt.show()
```

```python
plt.figure()
sns.countplot(x='campaign_type', data=df, hue='campaign_type', legend=False)
plt.title("Campaign Type Distribution")
plt.show()
```

```python
plt.figure()
sns.countplot(x='industry', data=df, hue='industry', legend=False)
plt.xticks(rotation=45)
plt.title("Industry Distribution")
plt.show()
```

```python
plt.figure()
sns.histplot(df['engagement'], kde=True)
plt.title("Engagement Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='platform', y='engagement', data=df)
plt.title("Engagement by Platform")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='followers', y='engagement', data=df)
plt.title("Followers vs Engagement")
plt.show()
```

```python
avg_engagement = df.groupby('platform')['engagement'].mean()

plt.figure()
plt.plot(avg_engagement.index, avg_engagement.values, marker='o')
plt.title("Average Engagement by Platform")
plt.xlabel("Platform")
plt.ylabel("Average Engagement")
plt.show()
```

```python
plt.figure()
sns.countplot(x='industry', hue='platform', data=df)
plt.xticks(rotation=45)
plt.title("Platform Usage Across Industries")
plt.show()
```

```python
plt.figure()
sns.barplot(
    x='campaign_type',
    y='engagement',
    hue='platform',
    data=df,
    errorbar=None
)
plt.title("Average Engagement by Campaign Type and Platform")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='followers', y='engagement', hue='platform', data=df)
plt.title("Followers vs Engagement by Platform")
plt.show()
```

```python
num_df = df[['engagement', 'followers']]
num_df.corr()
```

```python
plt.figure()
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Creating the pivot table & # Plot heatmap
pivot_engagement = df.pivot_table(
    values='engagement',
    index='platform',
    columns='campaign_type',
    aggfunc='mean'
)

plt.figure()
sns.heatmap(pivot_engagement, annot=True, cmap='YlGnBu')
plt.title("Average Engagement by Platform and Campaign Type")
plt.show()
```
